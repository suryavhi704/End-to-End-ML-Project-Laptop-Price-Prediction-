# Feature Engineering:

Feature Transformation:

```python
# Cheking for fromatting for object columns:
for col in df.columns:
  if df[col].dtypes == 'object':
    if df[col].nunique() <=5:
      print(df[col].value_counts())
      print('-'*60)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Label Encoding:
le_cols=['Screen','Touchscreen','IPSpanel','RetinaDisplay','CPU_company','PrimaryStorageType','SecondaryStorageType','GPU_company']
for col in le_cols:
  df[col]=le.fit_transform(df[col])
```

```python
df.head(4)
```

```python
# Target Encoding:
for col in df.columns:
  if col not in le_cols:
    if df[col].dtypes=='object':
      dictionary=df.groupby(col)['Price_euros'].mean().to_dict()
      df[col]=df[col].map(dictionary)
```

Feature Saling:

```python
x=df.drop(columns='Price_euros')
y=df['Price_euros']
```

```python
x.head()
```

```python
y.head()
```

```python
x.describe()
```

```python
# Robust Scaling:
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
x_scaled=scaler.fit_transform(x)
```

```python
# Creating scaled df:
new_df=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
new_df['target']=y
```

```python
new_df.head()
```

Feature Extraction:

```python
x_scaled=new_df.drop(columns='target')
y=new_df['target']
```

Feature Selection

```python
from sklearn.feature_selection import mutual_info_regression #feature scaling
imp=mutual_info_regression(x_scaled,y)
```

```python
scores=pd.Series(imp,index=x_scaled.columns)
```

```python
scores.sort_values(ascending=False)[:7] # top 7 columns
```

```python
imp_cols=scores.sort_values(ascending=False)[:7].index
imp_cols
```

```python
X=x_scaled[imp_cols]
Y=y
```

# Cross Validation:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=55)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Model Training:

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation:

```python
from sklearn.metrics import *
```

```python
y_pred=model.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
model.score(x_test,y_test)*100
```
